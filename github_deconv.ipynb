```python
import pandas as pd
import numpy as np
import scanpy as sc
import deconvolute as dc
```

```python
pbmc = sc.read_csv("20200323_pbmc_3K.csv")
bulk_data = pd.read_csv("Singer_AT_study_raw_counts.txt",delimiter='\t', index_col=0)
pbmc.obs_names_make_unique()
clusters = np.array(["T_Cells", "CD14+_Mono", "B_Cells", "FCGR3A+_Mono", "NK_Cells", "DC_Cells", "Platelets"])
```

```python
pbmc_norm = dc.normalize_cells(pbmc, var_genes = 4000, scatter_plot = True, pca_plot = True)
```
